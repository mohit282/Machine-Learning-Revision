```python
# before finding the correlation, check if the columns has null values 
df.isnull().sum()
```

> As we can see the rating column has 101 missing values (more then 10% of the column size), the rating of a smartphone depends on every other column, we cannot randomly fill its values, what we can do is that use every column and create a certain a logic which will help to fill those value, fortunately we have a special in-built method by sklearn library for doing this same task which is know as KNN Imputer. It will fill all the missing values based on what could be the best values by using other columns

> As of now, the relation of rating with price column is about 0.28, we will see that after filling the values using the KNN Imputer method, its correlation will also increase

```python
df.corr(numeric_only=True)['rating']
```

```python
# knn imputer
df.shape
```

```python
x_df = df.select_dtypes(include=['number']).drop(columns='price')
```

```python
x_df.shape
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=5)
```

```python
x_df_values = imputer.fit_transform(x_df)
```

```python
x_df_values
```

```python
x_df = pd.DataFrame(x_df_values,columns=x_df.columns)
```

```python
x_df['price'] = df['price']
```

```python
x_df.head()
```

```python
a = x_df.corr(numeric_only=True)['price'].reset_index()
```

```python
b = df.corr(numeric_only=True)['price'].reset_index()
```

```python
b.merge(a,on='index')
```
