```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import chi2
```

```python
# chi-square goodness-of-fit test
statistic = 19.6
degree_of_freedom = 4

p_value = 1 - chi2.cdf(statistic, degree_of_freedom)

print(p_value)

# Observed and expected frequencies
observed = np.array([420, 380, 410, 400, 410, 430, 350])
expected = np.array([400] * len(observed))  # Expected frequency of 400 for all categories

# Calculate chi-square statistic
chi_square_statistic = np.sum((observed - expected) ** 2 / expected)
chi_square_statistic
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

def plot_chi_square_distribution(df, x_max=20, num_points=1000):
    """
    Plot a chi-square distribution with specified degrees of freedom.

    Parameters:
    - df: int, degrees of freedom
    - x_max: float, maximum value of x-axis
    - num_points: int, number of points in x-axis for smooth curve
    """
    # Generate x values
    x = np.linspace(0, x_max, num_points)
    # Calculate chi-square distribution for given df
    y = chi2.pdf(x, df)
    
    # Plot the chi-square distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Chi-Square Distribution (df={df})', color='blue')
    plt.title(f'Chi-Square Distribution with {df} Degrees of Freedom')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage: plot chi-square distribution with 4 degrees of freedom
plot_chi_square_distribution(df=4)
```
