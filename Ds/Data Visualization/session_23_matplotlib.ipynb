### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.grid(True)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('datasets/sharma-kohli.csv')
print(batsman.head())

plt.plot(batsman['index'],batsman['V Kohli'])
# plt.plot(batsman['index'],batsman['RG Sharma'])
plt.grid(True)
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'],color='red',label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',label='Rohit Sharma')
plt.legend()
plt.grid(True)
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='blue')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# linestyle
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='p',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper left') #default loc='best'
plt.grid()
plt.show()
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
# Values of x and y can be limited
plt.ylim(0,75000) 
plt.xlim(2016,2020)
plt.grid(True)
```

# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('datasets/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(x=df['avg'],y=df['strike_rate'],color='red',marker='+')
plt.title('Avg runs and Strike Rate analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# size
tips = sns.load_dataset('tips')

# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10)
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o',marker='o')
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

# plt.bar(colors,children,color='black')
plt.bar(colors,children,color=['red','blue','green','yellow','pink'])
```

```python
# bar chart using data
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('datasets/batsman_season_record.csv')
df
```

```python
plt.bar(x=df['batsman'],height=df['2015'],width=0.9)
```

```python
#Multiple bar charts
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='brown',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='orange',label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='skyblue',label='2017')
plt.legend()
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# xticks
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.yticks(rotation=40)
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Histogram

image.png

- Univariate Analysis
- Numerical col
- Use case - Frequency Count

```python
# simple data

data = [32,45,56,38,10,63,66,68,15,27,61,19]

plt.hist(data,bins=[10,20,30,40,50,60,70],width=9,color='orange')
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [84,86,78,87,74]
subjects = ['eng','Physics','maths','Chemistry','Phy Edu']
plt.pie(data,labels=subjects,autopct='%0.1f%%',explode=[0.1,0.3,0,0,0],shadow=True)

plt.show()
```

```python
# dataset
df = pd.read_csv('datasets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
