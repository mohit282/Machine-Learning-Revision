```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('created_datasets/olx_new_data.csv')
df.head()
```

```python
df['PostedDate'].unique()
```

```python
ind = df[df['PostedDate'].str.contains('day')].index
```

```python
df.loc[ind,'PostedDate'] = 'Nov 2024'
```

```python
dates = ['Nov 2024', 'Sep-20']

import re

pd.to_datetime((re.findall(r'[a-zA-Z]{3}',dates[1])[0]+ ' 2024'), format='%b %Y').strftime('%Y-%m')
```

```python
def changeDate(date):
    month = re.findall(r'[a-zA-Z]{3}',date)[0]
    
    return month + '-24'


df['PostedDate'] = df['PostedDate'].apply(changeDate)
```

```python
age = 2024 - df['Year']
df.insert(3,'Age',age)
```

```python
df = df.drop(columns=['CarUrl'])
```

```python
maindf = pd.read_csv('created_datasets/used_car_dataset.csv')
maindf.head()
```

```python
df = pd.concat([maindf,df],axis=0)
df.shape
```

```python
df.sample(5)
```

```python
df.to_csv('used_cars_dataset_v2.csv',index=False)
```
