```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('olx_used_cars.csv')
df.head()
```

```python
df['PostedDate'].unique()
```

```python
ind = df[df['PostedDate'].str.contains('day')].index
```

```python
df.loc[ind,'PostedDate'] = 'Nov 2024'
```

```python
dates = ['Nov 2024', 'Sep-20']

import re

pd.to_datetime((re.findall(r'[a-zA-Z]{3}',dates[1])[0]+ ' 2024'), format='%b %Y').strftime('%Y-%m')
```

```python
def changeDate(date):
    month = re.findall(r'[a-zA-Z]{3}',date)[0]
    if month == 'Dec':
        return month + " 2023"
    
    else:
        return month + ' 2024'


df['PostedDate'] = df['PostedDate'].apply(changeDate)
```

```python
df.to_csv('used_cars_olx.csv',index=False)
```
